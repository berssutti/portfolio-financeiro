## Amostra de treinamento

### Definindo o periodo

```python
todayB = dt.date.today() - dt.timedelta(days=1*365)
yesterdayB = todayB - dt.timedelta(days=2*365)
print(todayB,yesterdayB)
```

```python
def historicalDay(coin,yesterdayB,todayB):
    df = yf.download(coin, start=str(yesterdayB), end=str(todayB)) 
    df=df[df.columns[:6]]
    return df
```

```python
df_qual3_trei, df_bpan4_trei, df_enev3_trei, df_bpac11_trei, df_pcar3_trei = pd.DataFrame(),pd.DataFrame(),pd.DataFrame(), pd.DataFrame(),  pd.DataFrame()

for symbol in tqdm(symbols):
    try: 
        
        if(symbol == symbols[0]):
            df = historicalDay(symbol,yesterdayB,todayB)
            df_qual3_trei = pd.DataFrame(index=df.index)
            df_qual3_trei['Open']=df.Open.values
            df_qual3_trei['High']=df.High.values
            df_qual3_trei['Low']=df.Low.values
            df_qual3_trei['Close']=df.Close.values
            df_qual3_trei.reset_index(inplace=True)
            df_qual3_trei['Date'] = df_qual3_trei['Date'].map(mdates.date2num)
            df_qual3_trei=df_qual3_trei[['Date', 'Open', 'High', 'Low', 'Close']]
        elif(symbol == symbols[1]):
            df = historicalDay(symbol,yesterdayB,todayB)
            df_bpan4_trei = pd.DataFrame(index=df.index)
            df_bpan4_trei['Open']=df.Open.values
            df_bpan4_trei['High']=df.High.values
            df_bpan4_trei['Low']=df.Low.values
            df_bpan4_trei['Close']=df.Close.values
            df_bpan4_trei.reset_index(inplace=True)
            df_bpan4_trei['Date'] = df_bpan4_trei['Date'].map(mdates.date2num)
            df_bpan4_trei=df_bpan4_trei[['Date', 'Open', 'High', 'Low', 'Close']]
        elif(symbol == symbols[2]):
            df = historicalDay(symbol,yesterdayB,todayB)
            df_enev3_trei = pd.DataFrame(index=df.index)
            df_enev3_trei['Open']=df.Open.values
            df_enev3_trei['High']=df.High.values
            df_enev3_trei['Low']=df.Low.values
            df_enev3_trei['Close']=df.Close.values
            df_enev3_trei.reset_index(inplace=True)
            df_enev3_trei['Date'] = df_enev3_trei['Date'].map(mdates.date2num)
            df_enev3_trei=df_enev3_trei[['Date', 'Open', 'High', 'Low', 'Close']]
        elif(symbol == symbols[3]):
            df = historicalDay(symbol,yesterdayB,todayB)
            df_bpac11_trei = pd.DataFrame(index=df.index)
            df_bpac11_trei['Open']=df.Open.values
            df_bpac11_trei['High']=df.High.values
            df_bpac11_trei['Low']=df.Low.values
            df_bpac11_trei['Close']=df.Close.values
            df_bpac11_trei.reset_index(inplace=True)
            df_bpac11_trei['Date'] = df_bpac11_trei['Date'].map(mdates.date2num)
            df_bpac11_trei=df_bpac11_trei[['Date', 'Open', 'High', 'Low', 'Close']]
        else:
            df = historicalDay(symbol,yesterdayB,todayB)
            df_pcar3_trei = pd.DataFrame(index=df.index)
            df_pcar3_trei['Open']=df.Open.values
            df_pcar3_trei['High']=df.High.values
            df_pcar3_trei['Low']=df.Low.values
            df_pcar3_trei['Close']=df.Close.values
            df_pcar3_trei.reset_index(inplace=True)
            df_pcar3_trei['Date'] = df_pcar3_trei['Date'].map(mdates.date2num)
            df_pcar3_trei=df_pcar3_trei[['Date', 'Open', 'High', 'Low', 'Close']]

    except:
        print("Erro",symbol)
amostra_treino = [df_qual3_trei,df_bpan4_trei,df_enev3_trei,df_bpac11_trei,df_pcar3_trei]
```

# Médias Moveis dos Ativos

```python
def ema(data, period=0, column='Close'):
        data['EMA' + str(period)] = data[column].ewm(ignore_na=False, min_periods=period, com=period, adjust=True).mean()    
        return data

for ativo in range(5):
    mean_short=12
    mean_long=26

    close_mean_short = amostra_treino[ativo].Close.rolling(window=mean_short).mean()
    close_mean_long = amostra_treino[ativo].Close.rolling(window=mean_long).mean()
    
    amostra_treino[ativo]['SMA_short'] = close_mean_short
    amostra_treino[ativo]['SMA_long'] = close_mean_long
    
    fig = plt.subplots(figsize=(20,12))
    plt.style.use('seaborn-whitegrid')

    ax1 = plt.subplot(211)
    candlestick_ohlc(ax1, amostra_treino[ativo].values, width=0.6, colorup='b', colordown='r',alpha=1)
    ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    ax1.plot(amostra_treino[ativo]['Date'], close_mean_short, color = 'green', label = 'Short')
    ax1.plot(amostra_treino[ativo]['Date'], close_mean_long, color = 'y', label = 'Long')
    ax1.plot(amostra_treino[ativo]['Date'], ema(amostra_treino[ativo],21)['EMA21'], color = 'gray', label = 'Ema')
    ax1.legend(loc='upper left',fontsize=20)

    ax1.set_xlabel("Dias",fontsize=15)
    ax1.set_ylabel('Preços OHLC [$]',fontsize=15)
    ax1.set_title('Media Movel ' + str( symbols[ativo]),fontsize=25) 

    ax2 = plt.subplot(212)
    ax2.plot(amostra_treino[ativo]['Date'], amostra_treino[ativo].Close, 'k.', label = 'Close')
    ax2.plot(amostra_treino[ativo]['Date'], close_mean_short, color = 'green', label = 'Short')
    ax2.plot(amostra_treino[ativo]['Date'], close_mean_long, color = 'y', label = 'Long')
    ax2.plot(amostra_treino[ativo]['Date'], ema(amostra_treino[ativo],21)['EMA21'], color = 'gray', label = 'Ema')
    ax2.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    ax2.legend(loc='upper left',fontsize=20)

    ax2.set_xlabel("Dias",fontsize=15)
    ax2.set_ylabel('Preço [$]',fontsize=15)
    ax2.set_title('Gráfico do Preço de Fechamento',fontsize=25) 
```

```python
def macd(data, period_long=26, period_short=12, period_signal=9):
    data['Macd'] = data['SMA_short'] - data['SMA_long']
    #Exponencial
    data['Macd_signal_line'] = data['Macd'].ewm(ignore_na=False, min_periods=0, com=period_signal, adjust=True).mean()
    return data

macd(trades)
trades['Macd_signal']=0
trades['Macd_signal'] = np.where(trades['Macd'] > trades['Macd_signal_line'], 1.0, 0.0)   

# Posição dos trades
trades['Macd_positions'] = trades['Macd_signal'].diff()

fig = plt.subplots(figsize=(25,25))
plt.style.use('seaborn-whitegrid')

ax1 = plt.subplot(621)
ax1.plot(trades['Close'], label='Close')

ax1.plot(trades.loc[trades.Macd_positions == 1.0].index, 
         trades.Close[trades.Macd_positions == 1.0],
         '^', markersize=10, color='m',label='Buy')
ax1.plot(trades.loc[trades.Macd_positions == -1.0].index, 
         trades.Close[trades.Macd_positions == -1.0],
         'v', markersize=10, color='k',label='Sell') 

ax1.set_ylabel("Preço [R$]",fontsize=25)
ax1.set_xlabel("Data",fontsize=25)
ax1.set_title("Médias",fontsize=25)
ax1.legend(loc='upper left',fontsize=25)

ax2 = plt.subplot(622)
ax2.plot(trades['Close'][-200:], label='Close')

ax2.set_ylabel("Preço [R$]",fontsize=25)
ax2.set_xlabel("Data",fontsize=25)
ax2.set_title("Médias",fontsize=25)

ax2.plot(trades[-200:].loc[trades[-200:].Macd_positions == 1.0].index, 
         trades[-200:].Close[trades[-200:].Macd_positions == 1.0],
         '^', markersize=10, color='m',label='Buy')
ax2.plot(trades[-200:].loc[trades[-200:].Macd_positions == -1.0].index, 
         trades[-200:].Close[trades[-200:].Macd_positions == -1.0],
         'v', markersize=10, color='k',label='Sell') 

ax2.legend(loc='upper left',fontsize=25)

ax3 = plt.subplot(623)
ax3.plot(trades['Macd'], 'k', label = 'MACD')
ax3.plot(trades['Macd_signal_line'], 'k.', label = 'SINAL MACD')
ax3.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax3.legend(loc='upper left',fontsize=25)

ax3.set_xlabel("Dias",fontsize=25)
ax3.set_ylabel('MACD',fontsize=25)
ax3.set_title('Indicador MACD',fontsize=25) 

ax4 = plt.subplot(624)
ax4.plot(trades['Macd'][-200:], 'k', label = 'MACD')
ax4.plot(trades['Macd_signal_line'][-200:], 'k.', label = 'SINAL')
ax4.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax4.legend(loc='upper left',fontsize=25)

ax4.set_xlabel("Dias",fontsize=25)
ax4.set_ylabel('MACD',fontsize=25)
ax4.set_title('Indicador MACD',fontsize=25) 

ax5 = plt.subplot(625)
ax5.plot(trades['Macd_signal'], label='signal')
ax5.set_ylabel("Signal",fontsize=25)
ax5.set_xlabel("Data",fontsize=25)
ax5.set_title("Sinal",fontsize=25)
ax5.legend(loc='upper left',fontsize=25)

ax6 = plt.subplot(626)
ax6.plot(trades['Macd_positions'], label='position')
ax6.set_ylabel("Position",fontsize=25)
ax6.set_xlabel("Data",fontsize=25)
ax6.set_title("Position",fontsize=25)
ax6.legend(loc='upper left',fontsize=25)

plt.tight_layout() 
```

## Contra Tendencia com Bandas de Bollinger 

```python
def bollinger_bands(data, trend_periods=20, deviation=2, close_col='Close'):
    data['Bol_bands_middle'] = data['Close'].ewm(ignore_na=False, min_periods=0, com=trend_periods, adjust=True).mean()
    for index, row in data.iterrows():
        s = data['Close'].iloc[index - trend_periods: index]
        sums = 0
        middle_band = data.at[index, 'Bol_bands_middle']
        for e in s:
            sums += np.square(e - middle_band)            
        std = np.sqrt(sums / trend_periods)
         #  d = 2
        upper_band = middle_band + (deviation * std)
        lower_band = middle_band - (deviation * std)
        # data.set_value(index, 'bol_bands_upper', upper_band)
        data.at[index, 'Bol_bands_upper']=upper_band 
       # data.set_value(index, 'bol_bands_lower', lower_band)
        data.at[index, 'Bol_bands_lower']=lower_band
    return data
```

```python
def sumper(per):
    persum=100*(np.exp(np.log1p(per/100).cumsum())-1)
    return persum
```
